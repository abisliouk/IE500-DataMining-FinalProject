### Performance analysis

```python
import matplotlib.pyplot as plt
import numpy as np

opacity = 0.5
bar_width = 0.35

# [Decision Tree, Random Forest, KNN, SVM, Naive Bayes, Baseline]
data_114_genres = [0.37, 0.61, 0.35, 0.33, 0.18, 0.03]
data_18_genres = [0.47, 0.66, 0.5, 0.45, 0.25, 0.09] # SVM 0.65 with 2000 authors

plt.figure(figsize=[15, 10])
x_coords = np.arange(len(data_114_genres))
bar_114_genres = plt.bar(x_coords + bar_width, data_114_genres, bar_width,
               align='center', alpha=opacity, color='darkorange', label='114 genres')
bar_20_genres = plt.bar(x_coords, data_18_genres, bar_width,
               align='center', alpha=opacity, color='g', label='18 genres')

# Add the sports for each bar and the height
for bar in bar_114_genres + bar_20_genres:
    height = bar.get_height()
    # Place height of each bar text at the top of the bar
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height:.2f}',
             ha='center', va='bottom')

plt.legend()
plt.ylabel('F1-score')
plt.title('Performance analysis for different ML methods', fontweight='bold')
plt.xticks(x_coords + bar_width/2.0,('Decision Tree', 'Random Forest', 'KNN', 'SVM', 'Naive Bayes', 'Baseline'))
plt.tight_layout()
plt.show()
```

### Correlation matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

og = pd.read_csv('data/spotify_data.csv', index_col=[0])
og = og.drop(columns=['track_id', 'album_name', 'track_genre', 'artists', 'track_name'])
explicit = og['explicit'].astype(int).to_numpy()
og['explicit'] = explicit
corr_og = og.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_og, annot=False, cmap='RdBu_r', fmt=".2f", linewidths=.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
